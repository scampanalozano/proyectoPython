```python
import pandas as pd
import matplotlib.pyplot as plt



data_path= "../data/procesados/datos_procesados.csv"
df = pd.read_csv(data_path)

print ("Estadisticas Basicas")
print (df.describe())

plt.hist(df["price"], bins=20, edgecolor="black")
plt.title("Distibución de precios")
plt.xlabel=("Precios")
plt.ylabel=("Frecuencia")
plt.show()

print ("Productos mas caros")
preciosAltos = df.nlargest(3, "price")
print (preciosAltos)


df['price'].hist(bins=30, color='orange', density=True, figsize=(10,5))
plt.xlabel=("Precios")
plt.show()




```
